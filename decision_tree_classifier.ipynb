```python
# Importing the libraries
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Importing the dataset
```

```python
dataset = pd.read_csv('Iris.csv')
print(dataset.head())
X = dataset.iloc[:,1:-1].values
y = dataset.iloc[:,-1].values
```

```python
# Encoding the categorical data
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
print(y)
```

```python
# Splitting the dataset into training and test set
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 1)
```

```python
print(X_train)
```

```python
print(y_train)
```

```python
print(X_test)
```

```python
print(y_test)
```

```python
# Feature Scaling
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
print(X_train)
```

```python
print(X_test)
```

```python
# Training the Decision Tree Classification model on the training set
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion = 'entropy')
classifier.fit(X_train, y_train)
```

```python
# Predicting the Test set results by comparing actual y (y_test) and predicted y(y_pred)
```

```python
y_pred = classifier.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test),1)),1))
```

```python
# Making the Confusion Matrix
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
# Calculating the accuracy of the model
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
# Defining the predict function
```

```python
def predictFlower(sepal_length, sepal_width, petal_length, petal_width):
    prediction = classifier.predict(sc.transform([[sepal_length, sepal_width, petal_length, petal_width]]))
    
    if(prediction == 0):
        predicted_flower = 'Iris-setosa'
        
    elif(prediction == 1):
        predicted_flower = 'Iris-versicolor'
        
    elif(prediction == 2):
        predicted_flower = 'Iris-virginica'
        
    print('Predicted Flower: ',predicted_flower)
```

```python
# Making predictions against new data
```

```python
sepal_length = 7.1
sepal_width = 2.5
petal_length = 5.7
petal_width = 1.6
predictFlower(sepal_length, sepal_width, petal_length, petal_width)
```

```python
sepal_length = 3.5
sepal_width = 1.7
petal_length = 3.2
petal_width = 1.2
predictFlower(sepal_length, sepal_width, petal_length, petal_width)
```
